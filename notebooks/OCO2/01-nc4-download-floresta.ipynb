```python
url_floresta = 'http://floresta.if.usp.br/ftp/public/MapbiomasAR/OCO2_L2'

pahts = [
    # # 'http://floresta.if.usp.br/ftp/public/MapbiomasAR/OCO2_L2/OCO2_data_2014', 
    # # 'http://floresta.if.usp.br/ftp/public/MapbiomasAR/OCO2_L2/OCO2_data_2015',
    # # 'http://floresta.if.usp.br/ftp/public/MapbiomasAR/OCO2_L2/OCO2_data_2016/OCO2_data_2016', 
    'http://floresta.if.usp.br/ftp/public/MapbiomasAR/OCO2_L2/OCO2_data_2017', 
    'http://floresta.if.usp.br/ftp/public/MapbiomasAR/OCO2_L2/OCO2_data_2018',
    # # 'http://floresta.if.usp.br/ftp/public/MapbiomasAR/OCO2_L2/OCO2_data_2019', 
    # # 'http://floresta.if.usp.br/ftp/public/MapbiomasAR/OCO2_L2/OCO2_data_2020',
    # # 'http://floresta.if.usp.br/ftp/public/MapbiomasAR/OCO2_L2/OCO2_data_2021'
]

def listFD(url):
    page = requests.get(url).text
    soup = BeautifulSoup(page, 'html.parser')
    return [url + '/' + node.get('href') for node in soup.find_all('a') if node.get('href')]

obj_list = []

for path in pahts:
    sub_paths = listFD(path)
    # print(sub_paths)
    for final_path in sub_paths:
        urls = listFD(final_path[:-1])
        urls = list(filter(lambda f: f.find(".nc4") >= 0, urls))
        # path = sub_url[:-1]
        if urls == []:
            False         
        else:    
            for url in urls:
                string = str(url)
                split = string.split('/')
                len_split = len(split)
                obj = {
                    'date':split[len_split -2],
                    'name':split[len_split -1],
                    'address':string,
                }
                obj_list.append(obj)

# print(obj_list)
```
