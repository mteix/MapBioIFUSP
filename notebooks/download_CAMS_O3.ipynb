Ozone from CAMS


```python
import cdsapi

c = cdsapi.Client()


my_years = [
            '2003', '2004', '2005',
            '2006', '2007', '2008',
            '2009', '2010', '2011',
            '2012', '2013', '2014',
            '2015', '2016', '2017',
            '2018', '2019', '2020',
            '2021', '2022',
        ]


for yr in my_years:

    c.retrieve(
        'cams-global-reanalysis-eac4-monthly',
        {
            'format': 'netcdf',
            'pressure_level': '1000',
            'variable': 'ozone',
            'model_level': '60',
            
            'month': [
                '01', '02', '03',
                '04', '05', '06',
                '07', '08', '09',
                '10', '11', '12',
            ],
            'year': yr,
            'product_type': 'monthly_mean',
            'area': [
                15.6, -85.84, -57.59,
                -34,
            ],
        },
        'O3_CAMS_' + str(yr) + '.nc')
```
